```python
data = pd.read_csv('fund_name20220826.csv',usecols = ['name'],encoding='gbk')
data
```

```python
symbol = data['name'].values.tolist()
symbol
```

```python
fund =[]
for idx in symbol:
    data1 = pro.fund_nav(ts_code=idx)
    data1 = pd.DataFrame(data1[['nav_date','adj_nav']])
    data1 = data1.drop_duplicates(['nav_date'])
    data1 = data1.sort_values('nav_date',axis=0,ascending=True)
    data1 = data1.rename(columns ={'adj_nav':idx})
            
    times = data1['nav_date'].astype(str)
    formatStr = '%Y%m%d'
    data1.index = pd.to_datetime(times,format=formatStr)
    data1 = data1.drop(columns='nav_date')
    #data1.to_csv('fund/%s.csv'%idx,index=True)
    time.sleep(6/8)
    fund.append(data1)
```

```python
def merge_data(data):
    n = len(data)
    df = pd.merge(data[0],data[1],left_index=True,right_index=True,how='inner')
    for i in range(2,n):
        df = pd.merge(df,data[i],left_index=True,right_index=True,how='inner')
    return df
```

```python
class father:
    def __init__(self, father_attribute="father"):
        self.father_attribute=father_attribute

class child_without_ini(father):
    pass

class child_with_ini(father):
    def __init__(self, child_attribute):
        self.child_attribute=child_attribute

class child_with_super(father):
    def __init__(self,father_attribute,super_attribute):
        self.super_attribute=super_attribute
        super().__init__(father_attribute)

test_class_without_ini=child_without_ini()
test_class_with_ini=child_with_ini('child')
test_class_with_super=child_with_super('new_father','super')
```
